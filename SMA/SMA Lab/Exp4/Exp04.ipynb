```python
nx.degree(G)
```

```python
nx.degree_centrality(G)
```

**Identify Most Influential Node:**


Compute degree centrality of each node and then sort nodes in descending order


```python
m_influential=nx.degree_centrality(G)
```

```python
for w in sorted(m_influential, key=m_influential.get, reverse=True):
  print(w, m_influential[w])
```

**Identify Most Important Connection:**

Compute eigenvector centrality of each node and sort in descending order

```python
nx.eigenvector_centrality(G)
```

```python
m_imp_link= nx.eigenvector_centrality(G)
```

```python
for w in sorted(m_imp_link, key=m_imp_link.get, reverse=True):
  print(w, m_imp_link[w])
```
